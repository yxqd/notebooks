```python
dir1 = os.path.join(small_sample, 'x0.375_E22.5')
Eg = np.load(os.path.join(dir1, 'Eg.npy'))
xg = np.load(os.path.join(dir1, 'xg.npy'))

res = np.load(os.path.join(dir1, 'res_sum.npy'))
print np.max(res)
max = np.max(res[res<1e10])
res[res>max]=max

plt.pcolormesh(xg, Eg, res.T)
plt.colorbar()
plt.clim(0, 10000)
```

# Compare

```python
plt.plot(E1, I1, '+', label='large sample')
plt.plot(E2, I2*2.85, '^', label='small sample')
plt.xlim(-12, 8)
plt.ylim(0, 1.5e6)
plt.legend(loc='upper left')
plt.title('Ei=100meV E=22.5meV')
```
