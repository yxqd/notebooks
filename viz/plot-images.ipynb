```python
import numpy as np
```

```python
x,y = np.meshgrid(np.arange(100), np.arange(100))
```

```python
a = (x-50)**2 + (y-50)**2
```

```python
plt.figure()
plt.imshow(a)
```

```python
plt.figure()
rows = 4
cols = 10
for i in range(rows):
    for j in range(cols):
        plt.subplot(rows, cols, cols*i+j+1)
        plt.imshow(a)
# plt.tight_layout()
```
